## Cluster strategies

```python
n_clusters = 5
phenos = rank_top_phenos('present',20,1,1)[:20]

# fetch emps
gfs = [gf_startdates[pheno] for pheno in phenos]
gfl = [gf_lengths[pheno] for pheno in phenos]
LAs = []
for pheno in phenos:
    LA = df_leaves.query(f'cvar=={pheno}').groupby('site').max()['LA'].mean()
    LAs.append(LA)

# compiled df_emps & standardize
df_emps = pd.DataFrame({'cvar':phenos, 'gfs': gfs, 'gfl': gfl, 'LA': LAs})
df_emps_std = StandardScaler().fit_transform(df_emps.iloc[:,1:])

# cluster
kmeans_mod = KMeans(n_clusters=n_clusters)
kmeans_mod.fit(df_emps_std)
y_kmeans = kmeans_mod.predict(df_emps_std)
df_emps['group'] = y_kmeans


fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot()
ax.scatter(gf_startdates, gf_lengths, alpha=0.5, s=200,
           facecolor='none', edgecolor='grey')
ax.scatter(df_emps.gfs, df_emps.gfl, c=df_emps.group, 
           cmap = Vivid_5.mpl_colormap, s=200, alpha=0.8)
ax.set_ylim(0,120)
ax.set_xlim(150,220)

for item, pheno in enumerate(phenos):
    ax.annotate(pheno, (gfs[item], gfl[item]), color='black')
```
