### Alternative visualization - synergistic (white) vs. antagonistic (gray) zones
- Synergistic: <br/>
Emergent properties have synergistic effects on achieving high performance (high yield & high yield stability) <br/>
- Antagonistic: <br/>
Emergent properties compromise one another; if certain emergent property helps achieve high yield, will also lead to low yield stability. Can't have the best of both worlds. 


```python
plot_pcc_emps_board('present')
```

```python
plot_pcc_emps_board_highlight(df_pcc_present, 'present')
```
