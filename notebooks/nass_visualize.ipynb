## Process & visualize USDA NASS maize data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shapereader
from palettable.colorbrewer.sequential import YlGn_8

from ideotype.nass_process import read_nass, summarize_nass
```

```python
# Read in & process USDA NASS data
df_nass = read_nass()
df_nass = df_nass.query('year>=1961 & year<=2005')
df_nass_summary = summarize_nass()
```

```python
# Average maize yield over years
plt.plot(df_nass.groupby('year').mean()['yield'], c='grey')
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1,1,1, projection=ccrs.AlbersEqualArea(
                     central_latitude=39.5, central_longitude=-98.35))

ax.set_extent([-123,-72,19,53])
ax.scatter(df_nass_summary.lon, df_nass_summary.lat, 
           transform=ccrs.PlateCarree(),
           c=df_nass_summary['area']*2.47,
           cmap=YlGn_8.mpl_colormap, vmax=100000, alpha=0.2)

ax.add_feature(cfeature.BORDERS, edgecolor='grey')
ax.add_feature(cfeature.COASTLINE, edgecolor='grey')
ax.add_feature(cfeature.STATES, edgecolor='grey', linewidth=0.5)

fig.subplots_adjust(wspace=0.1, hspace=-0.8)
```

```python
area = 10000
df_nass_grouped = df_nass_summary.groupby(['state_id','county_id']).mean()

fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1,1,1, projection=ccrs.AlbersEqualArea(
                     central_latitude=39.5, central_longitude=-98.35))
ax.set_extent([-123,-72,19,53])
df_sub = df_nass_grouped.query(f'area > {area/2.27}')
ax.scatter(df_sub.lon, df_sub.lat, 
           transform=ccrs.PlateCarree(), c='purple', alpha=0.2)

ax.add_feature(cfeature.BORDERS, edgecolor='grey')
ax.add_feature(cfeature.COASTLINE, edgecolor='grey')
ax.add_feature(cfeature.STATES, edgecolor='grey', linewidth=0.5)

fig.subplots_adjust(wspace=0.1, hspace=-0.8)
```

```python
df_nass_grouped = df_nass.groupby(['state_id','county_id']).mean()

fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1,1,1, projection=ccrs.AlbersEqualArea(
                     central_latitude=39.5, central_longitude=-98.35))
ax.set_extent([-123,-72,19,53])
ax.scatter(df_nass_grouped.lon, df_nass_grouped.lat, 
           transform=ccrs.PlateCarree(),
           c=df_nass_grouped['yield'],
           cmap=YlGn_8.mpl_colormap, vmin=0, vmax=10, alpha=0.2)
ax.add_feature(cfeature.BORDERS, edgecolor='grey')
ax.add_feature(cfeature.COASTLINE, edgecolor='grey')
ax.add_feature(cfeature.STATES, edgecolor='grey', linewidth=0.5)

fig.subplots_adjust(wspace=0.1, hspace=-0.8)
```

```python
from palettable.colorbrewer.sequential import YlGn_8
years = list(np.arange(1961,2006))

fig = plt.figure(figsize=(15,30))
for i in np.arange(45):
    ax = fig.add_subplot(9,5,i+1, projection=ccrs.AlbersEqualArea(
                         central_latitude=39.5, central_longitude=-98.35))
    ax.set_extent([-123,-72,19,53])
    df_year = df_nass[df_nass.year == years[i]]
    ax.scatter(df_year.lon, df_year.lat, 
               transform=ccrs.PlateCarree(),
               c=df_year.area,
               cmap=YlGn_8.mpl_colormap, vmin=1000, vmax=50000, alpha=0.5)
    ax.add_feature(cfeature.BORDERS, edgecolor='grey')
    ax.add_feature(cfeature.COASTLINE, edgecolor='grey')
    ax.add_feature(cfeature.STATES, edgecolor='grey', linewidth=0.5)

fig.subplots_adjust(wspace=0.1, hspace=-0.8)
```
