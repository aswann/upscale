## Analyzing outputs of 5 parameter sets ensemble simulaitons

Analyzing the first batch of ensemble simulations. <br/>
Simulation outputs stored: /home/disk/eos8/ach315/upscale/sims/opt5

```python
import dill
dill.load_session('sim_opt5_env.db')
```

```python
import numpy as np
import pandas as pd
import matplotlib.path as mpath
import matplotlib.pyplot as plt
import glob
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader
import xarray as xr
import datetime
import seaborn as sns

from palettable.colorbrewer.sequential import OrRd_6
from palettable.colorbrewer.sequential import YlGn_9
from palettable.colorbrewer.sequential import YlGnBu_8
from palettable.colorbrewer.sequential import RdPu_5
```

```python
# manual update of maizsim output column name
cols = ['jday', 'time', 'leaves', 'mature_lvs', 'drop_lvs', 'LA', 'LA_dead', 'LAI', 'RH', 
        'leaf_WP', 'PFD', 'solrad', 'temp_soil', 'temp_air', 'temp_can', 'ET_dmd', 'ET_suply',
        'Pn', 'Pg', 'resp', 'av_gs', 'LAI_sunlit', 'LAI_shaded', 'PFD_sunlit', 'PFD_shaded', 
        'An_sunlit', 'An_shaded', 'Ag_sunlit', 'Ag_shaded', 'gs_sunlit', 'gs_shaded', 'VPD',
        'N', 'N_dmd', 'N_upt', 'N_leaf', 'PCRL', 'dm_total', 'dm_shoot', 'dm_ear', 'dm_totleaf', 'dm_dropleaf', 
        'df_stem', 'df_root', 'roil_rt', 'mx_rootdept', 'available_water', 'soluble_c', 'note']
```

#### Reading in simulation outputs:

```python
years = np.arange(1961,1991)
cultivars = np.arange(0,5)
yrs = []
cvars = []
sites = []
dm_ears = []


for i in years:
    for j in cultivars:
        sims = glob.glob('/home/disk/eos8/ach315/upscale/sims/opt5/' + str(i) + '/var_' + str(j) + '/out1*.txt')
        for sim in sims:
            site = sim.split('/')[-1].split('_')[1]
            df = pd.read_csv(sim)
            df.columns = cols
            ear = df.dm_ear.max()
            yrs.append(i)
            cvars.append(j)
            sites.append(site)
            dm_ears.append(ear)
            
        print(i, j)
```

#### Compiling all sims into master dataframe:

```python
df = pd.DataFrame({'year': yrs, 'vars': cvars, 'site': sites, 'ear': dm_ears})
df_siteinfo = pd.read_csv("/home/disk/eos8/ach315/upscale/weadata/site_info.csv", index_col=0, dtype=str)
df = pd.merge(df, df_siteinfo, on='site')
df.head()
```

```python
# testing out sim outputs for year 1961
extent = [-123, -72, 19, 53]
fig, axs = plt.subplots(3,2, figsize=(10,10),
                        subplot_kw={'projection': 
                                    ccrs.AlbersEqualArea(central_latitude=39.5, 
                                                         central_longitude=-98.35)})

cvars = np.arange(0,5)
for ax, v in zip(axs.flatten(), cvars):
    df_v = df[(df.year==1961) & (df.vars==v)]
    ax.scatter(df_v.lon.astype(float)*-1, df_v.lat.astype(float), 
               c=df_v.ear, cmap=YlGn_9.mpl_colormap, vmin=0, vmax=30,
               transform=ccrs.PlateCarree(), alpha=1, s=50)
    ax.set_extent(extent)
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS)
    ax.add_feature(cfeature.STATES, edgecolor='grey', linewidth=0.5)
    ax.set_title('cultivar ' + str(v), fontweight='light')

lim = np.array([[0, 30]])
im1 = plt.imshow(lim, cmap=YlGn_9.mpl_colormap)
im1.set_visible("False")
cb_ax1 = fig.add_axes([0.82, 0.39, 0.01, 0.39]) # left, bottom, width, top
cb1 = plt.colorbar(im1, cax=cb_ax1)
cb1.ax.set_ylabel('mean simulated yield (g/plant)')
plt.subplots_adjust(right=0.8, wspace=0.01)
    
plt.savefig('/home/disk/eos8/ach315/upscale/figs/vars_1961.png', format='png', dpi=1200)
```

#### Summarizing results over all years:

```python
df_sitesum = df.groupby(['site', 'vars']).mean().reset_index()
df_sitesum = pd.merge(df_sitesum, df_siteyears, on='site')
df_sitesum.head()
```

```python
# testing out sim outputs for year 1961
fig, axs = plt.subplots(3,2, figsize=(10,10),
                        subplot_kw={'projection': 
                                    ccrs.AlbersEqualArea(central_latitude=39.5, 
                                                         central_longitude=-98.35)})

cvars = np.arange(0,5)
for ax, v in zip(axs.flatten(), cvars):
    df_v = df_sitesum[df_sitesum.vars==v]
    ax.scatter(df_v.lon.astype(float)*-1, df_v.lat.astype(float), 
               c=df_v.ear, cmap=YlGn_9.mpl_colormap, vmin=0, vmax=30,
               s=df_v.years.astype(int)*4.5, alpha=0.7, 
               transform=ccrs.PlateCarree())
    ax.set_extent(extent)
    ax.add_feature(cfeature.COASTLINE, edgecolor='grey')
    ax.add_feature(cfeature.BORDERS, edgecolor='grey')
    ax.add_feature(cfeature.STATES, edgecolor='grey', linewidth=0.5)
    ax.set_title('cultivar ' + str(v+1), fontweight='light')

lim = np.array([[0, 30]])
im1 = plt.imshow(lim, cmap=YlGn_9.mpl_colormap)
im1.set_visible("False")
cb_ax1 = fig.add_axes([0.82, 0.39, 0.01, 0.39]) # left, bottom, width, top
cb1 = plt.colorbar(im1, cax=cb_ax1)
cb1.ax.set_ylabel('mean simulated yield (g/plant)')
plt.subplots_adjust(right=0.8, wspace=0.01)
    
plt.savefig('/home/disk/eos8/ach315/upscale/figs/vars_allyears.png', format='png', dpi=800)
```
