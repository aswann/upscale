```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

from palettable.colorbrewer.sequential import Purples_9
from palettable.colorbrewer.sequential import YlGn_9
from palettable.colorbrewer.sequential import PuRd_5
from palettable.colorbrewer.diverging import BrBG_5
from palettable.colorbrewer.diverging import PuOr_5
```

```python
# 1. maizsim outputs
df_sims = pd.read_csv('/home/disk/eos8/ach315/upscale/data/sims_6105.csv', dtype={'site':'str'})

# 2. site & site-years
df_sites_all = pd.read_csv('/home/disk/eos8/ach315/upscale/weadata/site_summary.csv', dtype={'site': str})
siteyears = pd.read_csv('/home/disk/eos8/ach315/upscale/weadata/siteyears_filtered.csv', dtype={'site':str})
df_sites = df_sites_all[df_sites_all.site.isin(siteyears.site)]
df_sites.reset_index(inplace=True, drop=True)

# 3. weather
df_wea = pd.read_csv('/home/disk/eos8/ach315/upscale/weadata/wea_summary.csv', dtype={'site': 'str'}, index_col=0)
df_wea.reset_index(inplace=True, drop=True)

# 4. parameter
df_params = pd.read_csv('/home/disk/eos8/ach315/upscale/params/param_opt.csv')
df_params = df_params.drop(['rmax_ltar'], axis=1)
df_params['cvar'] = df_params.index

# 5. merge all
df_sims_params = pd.merge(df_sims, df_params, on='cvar')
df_sims_params_sites = pd.merge(df_sims_params, df_sites, on='site')
df_all = pd.merge(df_sims_params_sites, df_wea, on=['site','year'])

# 6. subset matured
df_matured = df_all[df_all.note == '"Matured"']

# 7. group by cvars or site
df_sims_sitemean = df_all.groupby(['site']).mean()
df_sims_yearmean = df_all.groupby(['year']).mean()
df_sims_cvarmean = df_all.groupby(['cvar']).mean()
```

```python
# set up climatology info
df_climate = pd.read_csv('/home/disk/eos8/ach315/upscale/weadata/wea_summary.csv', index_col=0)
df_climate_site = df_climate.groupby('site').mean()

temp_threshold = 18
vpd_threshold = 0.7

df_hotwet = df_climate_site[(df_climate_site.temp>temp_threshold) & (df_climate_site.vpd>vpd_threshold)]
df_hotdry = df_climate_site[(df_climate_site.temp>temp_threshold) & (df_climate_site.vpd<vpd_threshold)]
df_coolwet = df_climate_site[(df_climate_site.temp<temp_threshold) & (df_climate_site.vpd>vpd_threshold)]
df_cooldry = df_climate_site[(df_climate_site.temp<temp_threshold) & (df_climate_site.vpd<vpd_threshold)]

hotwet_sites = df_hotwet.index.tolist()
hotdry_sites = df_hotdry.index.tolist()
coolwet_sites = df_coolwet.index.tolist()
cooldry_sites = df_cooldry.index.tolist()

hotwet_sites = [str(site) for site in hotwet_sites]
hotdry_sites = [str(site) for site in hotdry_sites]
coolwet_sites = [str(site) for site in coolwet_sites]
cooldry_sites = [str(site) for site in cooldry_sites]
```

```python
def linear_mod(df, features, target, test_size=0.33):
    """
    Linear model that operates from DF based on features & target.

    Parameters
    ----------
    df : pd.DataFrame
        Dataframe to draw data for linear model.
    features : 
    target : 
    test_size : 
        Default as 0.33 - 33% of data set aside for testing.
    """

    X = df[features]
    y = df[target]
    
#    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size)
    mod = LinearRegression(fit_intercept=True)
#    mod.fit(X_train, y_train)
    mod.fit(X, y)
    y_pred = mod.predict(X)
    coefs = mod.coef_
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    return coefs, mse, r2
```

```python
def mod_r2(df, features, target, cvars=100):
    """
    Linear model R2 values for each cultivar.
    
    Parameters
    ----------
    df : 
    features : 
    target : 
    cvars : Integer
        Number of cultivars.
    """

    r2_list = []
    for cvar in np.arange(cvars):
        df_cvar = df[df.cvar==cvar]
        coefs, mse, r2 = linear_mod(df_cvar, features, target)
        r2_list.append(r2)
    return r2_list
```

#### R2 climatology linear mod

1. Year-to-year yield variance averaged over all sites

```python
features = ['vpd']
target = ['dm_ear']

df = df_matured
r2_list = mod_r2(df, features, target)

ear_mean = df_all.groupby('cvar').mean().dm_ear
ear_var = df_matured.groupby(['cvar','site']).var().reset_index().groupby('cvar').mean().dm_ear

fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot()
#ax.scatter(np.arange(len(r2_list)), r2_list, c=ear_mean, s=100, alpha=0.7, cmap='YlGn')

ax.scatter(ear_mean, r2_list, c=ear_var, s=100, alpha=0.7, cmap='PuRd')
for i in np.arange(100):
    ax.annotate(i, (ear_mean[i], r2_list[i]), size=8)
ax.set_xlabel('mean yield')
ax.set_ylabel('r2 - climatology mod')
```

2. Year-to-year yield variance broken down by climate space

```python
sites = [hotwet_sites, hotdry_sites, coolwet_sites, cooldry_sites]
texts = ['hot-wet', 'hot-dry', 'cool-wet', 'cool-dry']

#features = ['temp', 'vpd', 'precip', 'solrad']
features = ['temp']
target = ['dm_ear']

fig, axs = plt.subplots(2, 2, figsize=(14,7))

for site, text, ax in zip(sites, texts, axs.flatten()):
    df = df_matured
    r2_list = mod_r2(df, features, target)

    ear_var = df_matured.groupby(['cvar', 'site']).var().query(f'site=={site}').reset_index().groupby('cvar').mean().dm_ear
    ear_mean = df_matured.groupby(['cvar', 'site']).mean().query(f'site=={site}').reset_index().groupby('cvar').mean().dm_ear
#    ear_mean = df_matured.groupby('cvar').mean().dm_ear
    
    ax.scatter(ear_mean, r2_list, c=ear_var, s=120, alpha=0.7, cmap='PuRd')
    ax.set_xlim(0, 150)
    ax.set_ylim(0, 0.9)

    for cvar in np.arange(100):
        ax.annotate(cvar, (ear_mean[cvar], r2_list[cvar]), size=8, color='dimgrey')
    ax.set_xlabel('mean yield')
    ax.set_ylabel('r2 - climatology mod')
    ax.set_title(f'variance across years - {text}')    

fig.subplots_adjust(left=0.05, bottom=0.1, right=0.95, top=0.95, wspace=0.15, hspace=0.4)

#plt.savefig('/home/disk/eos8/ach315/upscale/figs/r2allclimatemod_var_cspace.png', format='png', dpi=800)
```
