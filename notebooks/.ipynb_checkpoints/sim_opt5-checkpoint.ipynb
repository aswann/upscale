```python
years = np.arange(1961,1991)
cultivars = np.arange(0,5)
yrs = []
cvars = []
sites = []
dm_ears = []


for i in years:
    for j in cultivars:
        sims = glob.glob('/home/disk/eos8/ach315/upscale/sims/opt/' + str(i) + '/var_' + str(j) + '/out1*.txt')
        for sim in sims:
            site = sim.split('/')[-1].split('_')[1]
            df = pd.read_csv(sim)
            df.columns = cols
            ear = df.dm_ear.max()
            yrs.append(i)
            cvars.append(j)
            sites.append(site)
            dm_ears.append(ear)
            
        print(i, j)
```
