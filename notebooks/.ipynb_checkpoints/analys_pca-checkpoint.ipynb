```python
# fig setup
fig = plt.figure(figsize=(16,10))

for item, top_pheno in enumerate(top_pheno_sorted[:20]):
    # top pheno
    pheno_num = top_pheno
        
    # sims
    df_sims_pheno = df_sims.query(f'cvar=={pheno_num}')[['year','site','yield','soil_texture']]
    df_sims_pheno.reset_index(drop=True, inplace=True)
    df_sims_pheno['soil_texture_num'] = list(df_sims_pheno.soil_texture.factorize()[0])

    # climate
    df_tempvpd_pheno= df_gseason_climate.query(f'cvar=={pheno_num}')[['year', 'site', 'pheno', 'temp_air', 'vpd']]
    df_wd_pheno = df_waterdeficit_sum.query(f'cvar=={pheno_num}')[['year','site','pheno','water_deficit_sum']]

    # merge climate
    df_all = pd.merge(df_tempvpd_pheno, df_wd_pheno)

    # Filter out climate during phenostage
    df_em = df_all[df_all['pheno'] == '"Emerged"']
    df_em = pd.merge(df_sims_pheno, df_em)
    df_ti = df_all[df_all['pheno'] == '"Tasselinit"']
    df_ti = pd.merge(df_sims_pheno, df_ti)
    df_td = df_all[df_all['pheno'] == '"Tasseled"']
    df_td = pd.merge(df_sims_pheno, df_td)
    df_sk = df_all[df_all['pheno'] == '"Silked"']
    df_sk = pd.merge(df_sims_pheno, df_sk)
    df_gf = df_all[df_all['pheno'] == '"grainFill"']
    df_gf = pd.merge(df_sims_pheno, df_gf)

    # construct pandas table to run PCA
    df = pd.DataFrame({#'soil_texture_num': df_em['soil_texture_num'],
                       'temp_em': list(df_em['temp_air']),
                       'temp_ti': list(df_ti['temp_air']),
                       'temp_td': list(df_td['temp_air']),
                       'temp_sk': list(df_sk['temp_air']),
                       'temp_gf': list(df_gf['temp_air']),
                       'vpd_em': list(df_em['vpd']),
                       'vpd_ti': list(df_ti['vpd']),
                       'vpd_td': list(df_td['vpd']),
                       'vpd_sk': list(df_sk['vpd']),
                       'vpd_gf': list(df_gf['vpd']),
                       'wd_em': list(df_em['water_deficit_sum']),
                       'wd_ti': list(df_ti['water_deficit_sum']),
                       'wd_td': list(df_td['water_deficit_sum']),
                       'wd_sk': list(df_sk['water_deficit_sum']),
                       'wd_gf': list(df_gf['water_deficit_sum'])})
    #                   'yield': list(df_sims_pheno['yield'])})

    # fill nan values with 0
    df.fillna(value=0, inplace=True)

    # visualize
    n = 8
    x = df
    x = StandardScaler().fit_transform(x)
    pca = PCA(n_components=n)
    principalComponents = pca.fit_transform(x)
    column_labels = [f'PC{comp+1}' for comp in np.arange(n)]
    df_pca = pd.DataFrame(data=principalComponents,
                          columns=column_labels)

    # visualizing pca output
    x_pos = np.arange(len(pca.components_[0]))
    bars = list(df.columns)
    ax = fig.add_subplot(4,5,item+1)
    ax.bar(x_pos, pca.components_[0], 
           color='grey', alpha=0.5)
    ax.set_ylim(-0.5, 0.5)
    ax.axhline(y=0, linewidth=0.5, color='grey')
    ax.annotate(top_pheno, (0, 0.35), fontweight='light', size=10)
    if 14 < item < 20:
        plt.xticks(x_pos, bars, fontweight='light', rotation=90, size=9)

fig.subplots_adjust(top=0.94, wspace=0.35, hspace=0.3)

#plt.savefig('/home/disk/eos8/ach315/upscale/figs/pcas_componentbars_top_performers.png', format='png', dpi=800)
```
